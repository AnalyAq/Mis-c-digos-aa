```python
#Na y Tetha
Na=np.arange(97,189)#Días desde el 6 de abril hasta el 6 de julio
teta=2*np.pi*Na/365
#Declinacion
dec = 0.409*np.sin(teta-1.39)

L=-12.0827
L= np.deg2rad (L)
L
#distancia T-S PROM ^2
c1=0.033523
c2=0.00128
c3=0.000739
c4=0.000099
dist=1.0001+c1*np.cos(teta)+c2*np.sin(teta)+c3*np.cos(2*teta)+c4*np.sin(2*teta)
#ANGULO HORARIO rad
H=np.arccos(-np.tan(L)*np.tan(dec))
#FOTOPERIODO
N = 2*(np.degrees(H)/15)
#QS
Qs=0.41*37.211*(dist)*((H*np.sin(L)*np.sin(dec))+(np.cos(L)*np.cos(dec)*np.sin(H)))
Qs = Qs[:len(df)]
N = N[:len(df)]  # Slice N to match the length of df
vt = df["vt"].values
df["Qs"]=Qs
df
#Calculo del (RAD GLOBAL (Q+q)i) Qi costa central
RG=Qs*(0.06+((0.64*vt)/N))
RG
df["RG"]=RG
df
```

```python
#CALCULO DEL ALBEDO
RI1=df["Ri1"]
RI2=df["Ri2"]
RI3=df["Ri3"]
RR1=df["Rr1"]
RR2=df["Rr2"]
RR3=df["Rr3"]

a=((RR1/RI1)+(RR2/RI2)+(RR3/RI3))/3
a
df["a"]=a
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Dato KcPROM
dias = [11, 38, 77, 113]
kc = [0.41, 0.87, 1.51, 0.83]

interpolacion = interp1d(dias, kc, kind='cubic', fill_value='extrapolate')
dias_interpolados = np.linspace(1, 120, 500)
kc_interpolados = interpolacion(dias_interpolados)

plt.figure(figsize=(10, 6))
plt.scatter(dias, kc, color='red', label='Datos Promedio')
plt.plot(dias_interpolados, kc_interpolados, label='Curva Interpolada', color='skyblue')
plt.axvline(x=25, color='gray', linestyle='--')
plt.text(10, 0.65, 'Etapa I', verticalalignment='center', color='gray')
plt.axvline(x=55, color='gray', linestyle='--')
plt.text(35, 0.65, 'Etapa II', verticalalignment='center', color='gray')
plt.axvline(x=100, color='gray', linestyle='--')
plt.text(73, 0.65, 'Etapa III', verticalalignment='center', color='gray')
plt.text(110, 0.65, 'Etapa IV', verticalalignment='center', color='gray')
plt.xlabel('Días del Ciclo de Cultivo')
plt.ylabel('Kc')
plt.title('Curva Interpolada de Kc a lo Largo del Ciclo de Cultivo de la papa Canchan')
plt.legend()
plt.grid(True)
plt.show()
```
